```python
import os
import cv2
import yaml
import pandas as pd
import random
import numpy as np
%matplotlib inline
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
yaml_path = os.path.join(path,'data.yaml')
Train_img_path = os.path.join(os.path.join(path,"train"),"images")
Val_img_path = os.path.join(os.path.join(path,"val"),"images")
save_path = os.path.join(path,"save_weights")
if os.path.isdir(save_path):
    os.mkdir(save_path)
```

```python
model = YOLO("yolov8m.pt")
model.train(data=yaml_path, epochs = 200, patience = 30, imgsz = 512,project=save_path)
```

```python
model = YOLO("C:/Users/USER/runs/detect/train84/weights/best.pt")
test_loc = os.path.join(path,"test")
results = model.predict(source = test_loc,conf=0.96,save=True,project=save_path)
result_list = []
for result in results:
    if (len(result.boxes.cls)==0):
        result_list.append(0)
    else:
        result_list.append(1)
```

```python
submit = pd.read_csv('./sample_submission.csv')
submit['label'] = result_list
display(submit.head())
submit.to_csv('./save/submit5.csv', index=False)
```
